### 3-10 네이버 영화 평점 기준 영화의 평점 변화 확인하기 ###

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import urlopen

url_base = 'http://movie.naver.com/'
url_sub = 'movie/sdb/rank/rmovie.nhn?sel=cur&date=20170806'

page = urlopen(url_base + url_sub)

soup = BeautifulSoup(page, "html.parser")
soup
```

```python
date = pd.date_range('2017-5-1', periods=100, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = 'http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}'
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, 'html.parser')
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0,end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie['point'] = movie['point'].astype(float)
movie.head()
```

```python
import numpy as np

movie_uniq = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_uniq.head()
```

```python
movie_best = movie_uniq.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == "노무현입니다"')
tmp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```

### 3-11 영화별 날짜 변화에 따른 평점 변화 확인하기 ###

```python
movie_pivot = pd.pivot_table(movie, index=['date'], columns=['name'], values=['point'])
movie_pivot.head()
```

```python
movie_pivot.columns = movie_pivot.columns.droplevel()
```
