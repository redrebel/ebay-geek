```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('03. best_sandwiches_listchicago2.csv', index_col=0)
df.head()
```

```python
gmaps_key = 'AIzaSyBWLT-kGunMsmGJ556ArSIPsyAQefftJHg'
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in df.index:
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n]+', '+'Chicago'
        gmaps_out = gmaps.geocode(target_name)
        location_output = gmaps_out[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng

mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping)
mapping
```
