```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open('./09. alice.txt').read()
alice_mask = np.array(Image.open('./09. alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add('said')
```

```python
import matplotlib.pyplot as plt
import platform

%matplotlib inline

import platform
path = "c:/windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc

if platform.system() == "Darwin":
   rc('font', family = 'AppleGothic')
elif platform.system() == "Windows":
   font_name = font_manager.FontProperties(fname=path).get_name()
   rc('font', family=font_name)
else:
   print("Unkonwn system.. sorry~~")
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', max_words=200, mask=alice_mask,
               stopwords = stopwords)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show
```

```python
text = open('./09. a_new_hope.txt').read()

text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('./09. stormtrooper_mask.png'))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("int")
stopwords.add("ext")
```

```python
wc = WordCloud(max_words = 1000, mask=mask, stopwords=stopwords,
              margin=10, random_state=1).generate(text)

default_color = wc.to_array()
```

```python
import random
def grey_color_func(word, font_size, position, orientation,
                   randow_state=None, **kwargs):
    return 'hsl(0,0%%, %d%%)' % random.randint(60,100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3),
          interpolation='bilinear')
plt.axis('off')
plt.show()
```
